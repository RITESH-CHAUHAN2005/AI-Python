```python
df = pd.read_csv("/content/Unclean Dataset.csv", encoding='latin1')
df.head()
```

```python
df.columns = df.columns.str.strip()
df_split = df['Student_ID'].str.split(' | ', expand=True)

df['Student_ID'] = pd.to_numeric(df_split[0].str.strip(), errors='coerce').astype('Int64')
df['First_Name'] = df_split[1].str.strip()
df['Last_Name'] = df_split[2].str.strip()
df['Age'] = pd.to_numeric(df_split[3].str.strip(), errors='coerce').astype('Int64')
df['Gender'] = df_split[4].str.strip()
df['Course'] = df_split[5].str.strip()
df['Enrollment_Date'] = pd.to_datetime(df_split[6].str.strip(), errors='coerce')
df['Total_Payments'] = pd.to_numeric(df_split[7].str.replace('$', '', regex=False).str.strip(), errors='coerce')
```

Now that we've extracted and assigned the values, let's look at the first few rows of the cleaned DataFrame `df` and its information to confirm the changes.
